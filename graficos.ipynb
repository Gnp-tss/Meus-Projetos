```python
import matplotlib.pyplot as plt
import numpy as np
x=np.linspace(0,5,50)
func=np.cos(x)
plt.plot(x,func, label= 'cos(x)', color='purple', linestyle='--', linewidth=2)
plt.title('Cosine Function')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.legend()
plt.grid("true")
plt.show()


```
